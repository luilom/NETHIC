```python
import pandas as pd
from scipy import spatial
```

```python
root = pd.read_pickle("../dataframes/single_categories/root.pkl")
```

```python
root["filename"].tolist()
```

```python
v1 = root[root["filename"] == "animal_product_10562587"]["vector"].tolist()[0]
v2 = root[root["filename"] == "cars_12983997"]["vector"].tolist()[0]
v3 = root[root["filename"] == "beverages_and_sugar_10261898"]["vector"].tolist()[0]
v4 = root[root["filename"] == "agricultural_structures_and_production_11075285"]["vector"].tolist()[0]
```

```python
print("V1-V2 = {}".format(1 - spatial.distance.cosine(v1,v2)))
print("V1-V3 = {}".format(1 - spatial.distance.cosine(v1,v3)))
print("V1-V4 = {}".format(1 - spatial.distance.cosine(v1,v4)))
print("V2-V3 = {}".format(1 - spatial.distance.cosine(v2,v3)))
print("V2-V4 = {}".format(1 - spatial.distance.cosine(v2,v4)))
print("V3-V4 = {}".format(1 - spatial.distance.cosine(v3,v4)))
```
